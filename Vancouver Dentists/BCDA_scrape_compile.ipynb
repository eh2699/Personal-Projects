```python
import pandas as pd
import glob

path =r'C:/Users/krist/Desktop/Datasets/dentistry/geocode/latlong' # use your path
allFiles = glob.glob(path + "/*.csv")
frame = pd.DataFrame()
list_ = []
for file_ in allFiles:
    df = pd.read_csv(file_,index_col=None, header=0)
    list_.append(df)
frame = pd.concat(list_)
```

```python
#----------Display all rows & change options mode
pd.set_option('display.max_rows', 500)
pd.options.mode.chained_assignment = None
```

```python
#----------Clean cols
cols = ['Name', 'Business Name', 'Phone', 'Address', 'Lat', 'Lng']
frame = frame[cols]
```

```python
frame.to_csv('dental_office_locations.csv')
```
